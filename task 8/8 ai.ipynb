```python
def minimax(depth, index, isMax, values, maxDepth):
    if depth == maxDepth:
        return values[index]
    if isMax:
        return max(minimax(depth+1, index*2, False, values, maxDepth),
                   minimax(depth+1, index*2+1, False, values, maxDepth))
    else:
        return min(minimax(depth+1, index*2, True, values, maxDepth),
                   minimax(depth+1, index*2+1, True, values, maxDepth))

# Different attributes
values = [10, 15, 5, 25, 3, 7, 9, 2]  # changed values
maxDepth = 3                          # changed depth
print("Optimal value with new data:", minimax(0, 0, True, values, maxDepth))
```
